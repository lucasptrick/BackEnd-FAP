##Exerccio 1
**Data por extenso.** Faa um programa que solicite a data de nascimento (dd/mm/aaaa) do usurio e imprima a data com o nome do ms por extenso.

```python
mes_extenso = ['janeiro', 'fevereiro', 'maro',
               'abril','maio','junho','julho','agosto',
               'setembro','outubro','novembro','dezembro']

dia = int(input('Digite o DIA de nascimento: '))
mes = int(input('Digite o MS de nascimento: '))
ano = int(input('Digite o ANO de nascimento: '))

mes = mes - 1
mes_escolhido = (mes_extenso[mes])

print(f'Sua data de nascimento com o MS por *extenso* : {dia} de {mes_escolhido} de {ano}')
```
